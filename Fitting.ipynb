```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

# Generate synthetic quadratic data
np.random.seed(42)
X = np.linspace(-3, 3, 30).reshape(-1, 1)
y = X**2 + np.random.normal(0, 1, X.shape)  # Quadratic function with noise

# Create models for Degree 1 (Underfitting) and Degree 10 (Overfitting)
models = {
    "Underfitting (Degree 1)": make_pipeline(PolynomialFeatures(1), LinearRegression()),
    "Overfitting (Degree 10)": make_pipeline(PolynomialFeatures(10), LinearRegression())
}
```

```python
plt.figure(figsize=(12, 5))

for i, (title, model) in enumerate(models.items()):
    model.fit(X, y)  # Train the model
    y_pred = model.predict(X)  # Get predictions

    # Plot results
    plt.subplot(1, 2, i + 1)
    plt.scatter(X, y, color='blue', label="Actual Data")
    plt.plot(X, y_pred, color='red', linewidth=2, label=title)
    plt.legend()
    plt.title(title)

plt.tight_layout()
plt.show()
```
