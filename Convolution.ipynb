```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load and convert image to grayscale
image = cv2.imread('/content/bldcongrats.jpg', cv2.IMREAD_GRAYSCALE)
# Define a simple edge detection kernel
kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
# Apply convolution
convoluted_image = cv2.filter2D(image, -1, kernel)
# Display original and convoluted images
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(convoluted_image, cmap='gray')
plt.title("Edge Detection")
plt.show()
```
