```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Generate synthetic customer data (Annual Income vs Spending Score)
np.random.seed(42)
income = np.random.randint(20, 100, 100) # Annual Income in thousands
spending_score = np.random.randint(1, 100, 100) # Spending score
df = pd.DataFrame({'Annual Income': income, 'Spending Score': spending_score})
print(df.head())
```

```python
# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)
# Visualization
plt.figure(figsize=(8, 5))
plt.scatter(df['Annual Income'], df['Spending Score'], c=df['Cluster'], cmap='viridis', alpha=0.7)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("K-Means Clustering - Customer Segmentation")
plt.colorbar(label="Cluster")
plt.show()
```
