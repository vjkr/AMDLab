```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Simulated data: Temperature (°C) vs Water Flow (L/min)
temperature = np.array([10, 15, 20, 25, 30, 35, 40]).reshape(-1, 1)
water_flow = np.array([5, 7, 9, 11, 13, 15, 18])
# Train Linear Regression model
model = LinearRegression()
model.fit(temperature, water_flow)
predicted_flow = model.predict(temperature)
# Plot data points and regression line
plt.scatter(temperature, water_flow, color='blue', label='Actual Data')
plt.plot(temperature, predicted_flow, color='red', linewidth=2, label='Regression Line')
plt.xlabel("Temperature (°C)")
plt.ylabel("Water Flow (L/min)")
plt.title("Linear Regression: Temperature vs Water Flow")
plt.legend()
plt.grid()
plt.show()
```
