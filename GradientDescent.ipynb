```python
import numpy as np
import matplotlib.pyplot as plt
# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 1) * 10 # Feature
y = 5 * X + 3 + np.random.randn(100, 1) # Linear relationship with noise
# Initialize parameters
m, b = 0, 0 # Initial slope & intercept
learning_rate = 0.01
epochs = 100
# Gradient Descent
for _ in range(epochs):
  y_pred = m * X + b
  error = y_pred - y
  m -= learning_rate * (2 / len(X)) * np.sum(error * X)
  b -= learning_rate * (2 / len(X)) * np.sum(error)
print(f"Final slope (m): {m}, Final intercept (b): {b}")
```

```python
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, m * X + b, color='red', label="Gradient Descent Line")
plt.xlabel("X Feature")
plt.ylabel("Target y")
plt.title("Gradient Descent for Linear Regression")
plt.legend()
plt.show()
```
